```python
def get_data(data, q, c, i, a):
    q_n = qualities_dict[q]
    c_n = compression_types_dict[c]
    i_n = images_dict[i]
    k_n = accuracies_dict[a]
    return data[q_n, c_n, i_n, k_n]
```

```python
def get_rank_data(f_data, r_data, q, c, i):
    q_n = qualities_dict[q]
    c_n = compression_types_dict[c]
    i_n = images_dict[i]
    k_n = accuracies_dict[r_data[i]]
    r_l = f_data[q_n, c_n, i_n, k_n]
    return sorted(f_data[q_n, c_n, i_n], reverse = True).index(r_l) + 1
```

```python
def cr(imageA, imageB):
    a = os.path.getsize(imageA)
    b = os.path.getsize(imageB)
    return b/a
```

```python
def mse(imageA, imageB, write_result=False):
    with WandImage(filename=imageA) as a:
        with WandImage(filename=imageB) as b:
            result_image, result_metric = a.compare(b, metric="mean_squared")
            if write_result == True:
                with result_image:
                    result_image.save(filename='diff.JPEG')
    return result_metric
```

```python
def psnr(imageA, imageB, write_result=False):
    with WandImage(filename=imageA) as a:
        with WandImage(filename=imageB) as b:
            result_image, result_metric = a.compare(b, metric="peak_signal_to_noise_ratio")
            if write_result == True:
                with result_image:
                    result_image.save(filename='diff.JPEG')
    return result_metric
```

```python
def ssim(imageA, imageB, write_result=False):
    with WandImage(filename=imageA) as a:
        with WandImage(filename=imageB) as b:
            result_image, result_metric = a.compare(b, metric="structural_similarity")
            if write_result == True:
                with result_image:
                    result_image.save(filename='diff.JPEG')
    return result_metric
```
